```python
#We will be examining the different beverages Starbucks has to offer and finding the most healther alternative
#based on the nutrients provided


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

drinks = pd.read_csv('starbucks-menu-nutrition-drinks.csv')
food = pd.read_csv('starbucks-menu-nutrition-food.csv')
menu = pd.read_csv('starbucks_drinkMenu_expanded.csv')
```

```python
drinks = drinks.rename(columns = {'Unnamed: 0':'Drink', 'Protein':'Protein (g)', 'Sodium':'Sodium (g)'})
```

```python
drinks = drinks.replace('-', np.nan)
drinks = drinks.dropna()
```

```python
drinks[['Calories', 'Fat (g)', 'Carb. (g)', 'Fiber (g)', 'Protein (g)', 'Sodium (g)']] = \
drinks[['Calories', 'Fat (g)', 'Carb. (g)', 'Fiber (g)', 'Protein (g)', 'Sodium (g)']].astype(float)
```

```python
drinks.describe()
```

```python
food = food.rename(columns = {'Unnamed: 0':'Food'})
food = food.replace('-', np.nan)
food = food.dropna()
```

```python
food.describe()
```

```python
menu.info()
```

```python
menu.columns = menu.columns.str.replace(' ', '')
menu.columns = menu.columns.str.replace('%', '')
```

```python
menu = menu.rename(columns = {'Beverage_category':'Beverage Category', 'Beverage_prep':'Beverage Prep'})
```

```python
menu = menu.replace(['-', 'varies', 'Varies'], np.nan)
```

```python
menu = menu.dropna()
```

```python
menu.info()
```

```python
menu[['Caffeine(mg)']] = \
menu[['Caffeine(mg)']].astype(float)
```

```python
menu.hist(figsize=(20, 20))
plt.show()
```

```python
_ = plt.figure(figsize=(10, 10))
_ = sns.barplot(y='Beverage Category', x='Sugars(g)', orient='h', data=menu)
_ = plt.show()
```

```python
_ = plt.figure(figsize=(15, 15))
_ = sns.scatterplot(y='Caffeine(mg)', x='Calories', hue='Beverage Category', data=menu, s=100)
_ = plt.show()
```

```python
menu.sort_values('Calories')
```

```python
menu = menu[['Beverage Category', 'Calories', 'Cholesterol(mg)','Sugars(g)', 'Caffeine(mg)']]
```

```python
_ = plt.figure(figsize=(10, 10))
_ = sns.scatterplot(y='Caffeine(mg)', x='Sugars(g)', hue='Beverage Category', data=menu, s=100)
_ = plt.show()
```

```python
_ = plt.figure(figsize=(10, 10))
_ = sns.scatterplot(y='Cholesterol(mg)', x='Sugars(g)', hue='Beverage Category', data=menu, s=100)
_ = plt.show()
```
